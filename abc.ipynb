```python
def predict_cluster(new_data, kmeans_model=kmeans, numerical_columns=numerical_columns, cluster_mapping=cluster_mapping):
    """
    Predict the cluster for new data using the trained KMeans model.
    
    Parameters:
    - new_data: DataFrame containing the new data points
    - kmeans_model: Trained KMeans model
    - numerical_columns: List of numerical columns used for clustering
    - cluster_mapping: Dictionary mapping cluster labels to descriptions
    
    Returns:
    - DataFrame with the predicted cluster and cluster description
    """
    # Ensure new_data contains the required numerical columns
    new_data = new_data[numerical_columns]
    
    # Predict the cluster for the new data
    new_data['cluster'] = kmeans_model.predict(new_data)
    
    # Map the cluster labels to descriptions
    new_data['cluster_description'] = new_data['cluster'].map(cluster_mapping)
    
    return new_data

# Example usage:
# new_data = pd.DataFrame({
#     'flat_price': [50000, 75000],
#     'fav_food': [3, 2],
#     'eating_out': [4, 5],
#     'income': [100000, 150000],
#     'location_Agripada': [0, 0],
#     'location_Andheri': [1, 0],
#     'location_Bandra': [0, 1],
#     # ... other location columns ...
#     'location_others': [0, 0]
# })
# predicted_clusters = predict_cluster(new_data, kmeans, numerical_columns, cluster_mapping)
# print(predicted_clusters)
```
